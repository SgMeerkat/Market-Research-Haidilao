```python

# Select relevant columns for correlation analysis
selected_columns = [
    'Please indicate your age',
    'Which gender do you identify with?',
    'Please indicate your ethnicity.',
    'Which best describes your current profession?',
    'What is your average income (per year)?',
    'What is your relationship status?',
    'How often do you visit restaurants weekly?',
    'What is your average per person bill amount at a restaurant visit?',
    'When visiting a restaurant, what is your average group size?',
    'How often do you order food at home weekly?',
    'What is your average bill amount for food delivery at home?',
    'What is your favourite cuisine?',
    'Which dining experience do you prefer?',
    'Which factor is the most important for you when you are eating out in a restaurant?',
    'How likely are you to recommend a restaurant on social media after dining there?',
    'Do you know what a hot pot is?',
    'Do you like Asian cuisine?',
    'Do you prefer authentic Asian Cuisine over Asian fast food? ',
    'Have you ever tried eating hot pot?',
    'After learning about hot pot cuisine, how likely are you to try it?',
    'How likely are you to try out the experience of dining at Haidilao (including open kitchen, dancing waitresses, and manicures)?',
    'In which location did you last eat Asian cuisine?',
    'If a new Haidilao restaurant would open in Dublin, in which area would you visit it the most?',
    'After learning about Haidilao, how likely are you to try out Haidilao, if it opened in an area near you?',
    'How likely are you to visit a restaurant after seeing it on social media?'
]

# Create a new DataFrame with only the selected columns
selected_df = df[selected_columns]

# Encode categorical variables
encoded_df = pd.get_dummies(selected_df)

# Compute correlation matrix
correlation_matrix = encoded_df.corr()

# Plot correlation matrix using Seaborn
plt.figure(figsize=(30, 28))  # Increase size of the heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5, annot_kws={"size": 8})  # Adjust font size of annotations
plt.title('Correlation Matrix')
plt.show()
```

```python
# Correlation heatmap for numerical variables
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
plt.figure(figsize=(12, 8))
sns.heatmap(df[numerical_cols].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()

# Boxplot for numerical variables
for col in numerical_cols:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=col, data=df, orient='v')
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Select numerical columns for clustering
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
data_for_clustering = df[numerical_cols]

# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data_for_clustering)

# Choose the number of clusters
num_clusters = 5

# Fit KMeans clustering algorithm to the data
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans.fit(scaled_data)

# Add cluster labels to the original dataframe
df['Cluster'] = kmeans.labels_

# Analyze the clusters
cluster_means = df.groupby('Cluster').mean()
print(cluster_means)

# Visualize the clusters
for col in numerical_cols:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x='Cluster', y=col, data=df)
    plt.title(f'Cluster {col} Distribution')
    plt.show()
```

Cluster 0: This cluster represents individuals who are highly likely to recommend a restaurant on social media after dining there. They are also very likely to try out hot pot cuisine and the experience of dining at Haidilao. Additionally, they are highly inclined to visit a restaurant after seeing it on social media.

Cluster 1: This cluster consists of individuals who are moderately likely to recommend a restaurant on social media and try out hot pot cuisine. However, they are less likely to try the experience of dining at Haidilao and visit a restaurant after seeing it on social media compared to Cluster 0.

Cluster 2: Individuals in this cluster have mixed responses. They are moderately likely to try out hot pot cuisine and the experience of dining at Haidilao. However, they are less likely to recommend a restaurant on social media and visit a restaurant after seeing it on social media compared to Clusters 0 and 1.

Cluster 3: This cluster represents individuals who are highly likely to try out hot pot cuisine and Haidilao if it opened in their area. They are also likely to recommend a restaurant on social media and visit a restaurant after seeing it on social media.

Cluster 4: Individuals in this cluster are less likely to recommend a restaurant on social media, try out hot pot cuisine, the experience of dining at Haidilao, and visit a restaurant after seeing it on social media compared to other clusters.

```python
# Assuming you have a DataFrame named 'df' with columns 'Please indicate your age' and 'How often do you visit restaurants weekly?'
# Define a mapping dictionary to convert string values to numeric values
visit_frequency_mapping = {
    'None or rarely (0 - less than 1)': 0,
    'Rarely (less than once a month)': 1,
    'Sometimes (1-3 times a month)': 2,
    'Often (once a week or more)': 3,
    'Frequently (3-5 a week)': 4,
    'Routinely (more than 5 times a week)': 5
}

# Convert the string values to numeric values
df['Visit Frequency'] = df['How often do you visit restaurants weekly?'].map(visit_frequency_mapping)

# Group the data by age and calculate the average visit frequency
avg_spend_by_age = df.groupby('Please indicate your age')['Visit Frequency'].mean()

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(avg_spend_by_age.index, avg_spend_by_age.values, marker='o', linestyle='-')
plt.title('Average Visit Frequency by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Visit Frequency')
plt.grid(True)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()

```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns

# Select numerical columns for clustering
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
data_for_clustering = df[numerical_cols]

# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data_for_clustering)

# Choose the number of clusters
num_clusters = 3

# Fit KMeans clustering algorithm to the data
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans.fit(scaled_data)

# Add cluster labels to the original dataframe
df['Cluster'] = kmeans.labels_

# Analyze the clusters
cluster_means = df.groupby('Cluster').mean()
print(cluster_means)

# Visualize the clusters
for col in numerical_cols:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x='Cluster', y=col, data=df)
    plt.title(f'Cluster {col} Distribution', fontsize=14)  # Reduced text size
    plt.show()
```
